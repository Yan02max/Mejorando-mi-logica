```python
import numpy as np

precios = np.array([100, 250, 400, 150, 300])

#Aqui aumentamos el 10% de cada precio
precios_aumento = precios * 1.10
print(precios_aumento)
```

```python
precios_total = np.sum(precios)
promedio_precios = np.mean(precios)
valor_maximo = np.max(precios)
valor_minimo = np.min(precios)
valor_mayor_250 = np.sum(precios > 250)

# Impreciones
print(f"El precio total es: {precios_total}")
print(f"El promedio de los precios es: {promedio_precios}")
print(f"El valor maximo es: {valor_maximo}")
print(f"El valor minimo es: {valor_minimo}")
print(f"Los precios mayor a 250 son:{valor_mayor_250}")
```

```python
costos = np.array([50, 120, 200, 350, 100])

costos_altos = costos[costos > 100]
cantidad = np.sum(costos_altos > 100)
print(f"La cantidad de costos es: {cantidad}")
print(f"Los costos altos son: {costos_altos}")
```
